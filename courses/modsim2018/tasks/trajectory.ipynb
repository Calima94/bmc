```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
v = 30
a = 30
g = (a*math.pi)/180
sin = math.sin(g)
vy = v*sin
print(sin)
print(vy)
```

```python
cos = math.cos(g)
vx = v*cos
print(cos)
print(vx)
```

```python
t = np.arange(0,11,0.1)

def h(t):
    return 0 + vy*t + (1/2)*(-9.81)*(t**2)

def x(t):
    return 0 + vx*t

plt.plot(x(t),h(t))
plt.ylabel('height (m)')
plt.xlabel('distance (m)')
plt.grid()
plt.show()
```
